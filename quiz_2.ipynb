```python
import numpy as np 
import matplotlib.pyplot as plt

# Generate 40 numbers between 10 and 10 
x=np.linspace(-10, 10, 40)

# Parabola function
y = x**2

#Plot the function 
plt.plot(x, y)

#Set labels and title 
plt.xlabel("x-axis") 
plt.ylabel('y-axis') 
plt.title("Parabola")

# Display it 
plt.show()
```

```python
# question 15

def getNumber(x, y):
    '''This function finds something from two input values'''
    if x <= y:
        return x
    else:
        return y

print(getNumber.__doc__)
m = getNumber(5, 5)
print(m)
```

```python
data = '''1.23, 2.15, 3.11
1.21, 1.13, 1.46
2.13, 3.11, 2.12
'''

f = open('data.txt', 'w')
f.write(data)
f.close()

f = open('data.txt', 'r')
while True:
    line = f.readline()
    line = line.replace('\n', '')
    print(line)
    if len(line) == 0:
        break
f.close()
```
